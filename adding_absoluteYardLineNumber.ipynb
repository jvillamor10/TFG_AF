```python
weeks = [week1,week2,week3,week4,week5,week6,week7,week8,week9,week10,week11,week12,week13,week14,week15,week16,week17]
```

```python
for index,row in plays[plays["absoluteYardlineNumber"].isnull()].iterrows():
    week = row["id"].split(":")
    plays.loc[plays['id']==row["id"], ["absoluteYardlineNumber"]] = round(weeks[int(week[2])-1][(weeks[int(week[2])-1]["id"]==row["id"]) & (weeks[int(week[2])-1]["displayName"]=="Football")]["x"].values[0])
```
