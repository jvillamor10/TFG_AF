```python
for play in special_team_plays:
    id_parts = play.split(":")
    week = id_parts[2]
    plays.drop(plays[plays["id"]==play].index,inplace=True)
    weeks[int(week)-1].drop(weeks[int(week)-1][weeks[int(week)-1]["id"]== play].index,inplace=True)
```

```python
for week in weeks:
    week.reset_index(inplace=True)
```

```python
week_number = 1
for week in weeks:
    week.to_csv("processed_data/remove_st_nbt/week{}.csv".format(week_number))
    week_number+=1
```
