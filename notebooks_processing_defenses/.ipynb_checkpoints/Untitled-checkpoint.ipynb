```python
for play in all_plays["id"].unique():
    yFootball = all_plays[(all_plays["id"]==play)&(all_plays["displayName"]=="Football")]["y"].values[0]
    
    qb = all_plays[(all_plays["id"]==play)&(all_plays["position"] == "QB")]
    
    if len(qb) == 1 and abs(yFootball - qb["y"].values[0]) > 3:
        print(play)
        
```

```python
for play in all_plays["id"].unique():
    yFootball = all_plays[(all_plays["id"]==play)&(all_plays["displayName"]=="Football")]["y"].values[0]
    
    if yFootball > 53.3 or yFootball < 0:
        print(play)
```
