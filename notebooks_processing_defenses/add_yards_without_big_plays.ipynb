```python
import pandas as pd
```

```python
pd.set_option('display.max_columns',None)
```

```python
plays = pd.read_csv("../processed_data/clean/plays.csv",index_col=0)
plays.head()
```

```python
static_defense = pd.read_csv("../transformed_data/static_defenses.csv",index_col=0)
static_defense.head()
```

```python
def obtainYards(playId):
    return plays[plays["id"]==playId]["playResult"].values[0]
```

```python
static_defense["playResult"] = static_defense["id"].apply(obtainYards)
```

```python
static_defense.head()
```

```python
def replacePoint(number):
    return str(number).replace(".",",")
```

```python
static_defense.to_csv("../transformed_data/static_defenses_yards.csv")
```

```python
static_defense["defenseArea"] = static_defense["defenseArea"].apply(replacePoint)
static_defense["defenseAreaCoverDefenders"] = static_defense["defenseAreaCoverDefenders"].apply(replacePoint)
static_defense["width"] = static_defense["width"].apply(replacePoint)
static_defense["height"] = static_defense["height"].apply(replacePoint)
static_defense["differenceOffenseVsDefenseWidth"] = static_defense["differenceOffenseVsDefenseWidth"].apply(replacePoint)
static_defense["HeightByWeightDeep"] = static_defense["HeightByWeightDeep"].apply(replacePoint)
static_defense["HeightByWeightHook"] = static_defense["HeightByWeightHook"].apply(replacePoint)
static_defense["HeightByWeightCurl"] = static_defense["HeightByWeightCurl"].apply(replacePoint)
static_defense["HeightByWeightFlat"] = static_defense["HeightByWeightFlat"].apply(replacePoint)
static_defense["WeightByArea"] = static_defense["WeightByArea"].apply(replacePoint)
static_defense["density"] = static_defense["density"].apply(replacePoint)
static_defense["densityNoLine"] = static_defense["densityNoLine"].apply(replacePoint)
static_defense["densityNoLine"] = static_defense["densityNoLine"].apply(replacePoint)
static_defense["densityInsidePoints"] = static_defense["densityInsidePoints"].apply(replacePoint)
static_defense["densityInsidePointsNoLine"] = static_defense["densityInsidePointsNoLine"].apply(replacePoint)
```

```python
static_defense.to_csv("../transformed_data/static_defenses_yards_bigml.csv")
```
