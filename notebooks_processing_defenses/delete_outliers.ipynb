```python
import pandas as pd
import pickle
```

```python
pd.set_option('display.max_columns',None)
```

```python
static_defenses = pd.read_csv("../transformed_data/static_defenses.csv",index_col=1)
static_defenses
```

```python
#load dataset
pickle_file = open('../resources/static_defenses_clusters/dbscan/outliers_minpts10.pickle','rb')
outliers = pickle.load(pickle_file)
pickle_file.close()
```

```python
list(outliers)
```

```python
new_df = static_defenses.copy()
```

```python
for play in list(outliers):
    new_df.drop(play,inplace=True)
```

```python
new_df
```

```python
new_df.to_csv("../transformed_data/static_defenses_without_outliers.csv")
```

```python
new_df.drop(["Unnamed: 0"],axis=1,inplace=True)
new_df.head()
```

```python
def replacePoints(number):
    return str(number).replace(".",",")
```

```python
new_df["defenseArea"] = new_df["defenseArea"].apply(replacePoints)
new_df["defenseAreaCoverDefenders"] = new_df["defenseAreaCoverDefenders"].apply(replacePoints)
new_df["width"] = new_df["width"].apply(replacePoints)
new_df["height"] = new_df["height"].apply(replacePoints)
new_df["differenceOffenseVsDefenseWidth"] = new_df["differenceOffenseVsDefenseWidth"].apply(replacePoints)
new_df["differenceOffenseVsDefenseStrongSide"] = new_df["differenceOffenseVsDefenseStrongSide"].apply(replacePoints)
new_df["HeightByWeightDeep"] = new_df["HeightByWeightDeep"].apply(replacePoints)
new_df["HeightByWeightHook"] = new_df["HeightByWeightHook"].apply(replacePoints)
new_df["HeightByWeightCurl"] = new_df["HeightByWeightCurl"].apply(replacePoints)
new_df["HeightByWeightFlat"] = new_df["HeightByWeightFlat"].apply(replacePoints)
new_df["WeightByArea"] = new_df["WeightByArea"].apply(replacePoints)
new_df["density"] = new_df["density"].apply(replacePoints)
new_df["densityNoLine"] = new_df["densityNoLine"].apply(replacePoints)
new_df["densityInsidePoints"] = new_df["densityInsidePoints"].apply(replacePoints)
new_df["densityInsidePointsNoLine"] = new_df["densityInsidePointsNoLine"].apply(replacePoints)
```

```python
new_df
```

```python
new_df.to_csv("../transformed_data/static_defenses_without_outliers_bigml.csv")
```
