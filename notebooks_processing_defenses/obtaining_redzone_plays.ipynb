La zona profunda empieza después de 15 yardas de la línea de scrimmage. Estaría interesante empezar con defensas que no estén en la red zone (las últimas 20 yardas hasta la línea de anotación).

La red zone es cuando el equipo atacante está dentro de las últimas 20 yardas de su oponente.

```python
plays["redZone"] = ""
```

```python
plays.head()
```

```python
for index,row in plays.iterrows():
    playDirection = defenses[defenses["id"]==row["id"]]["playDirection"].values[0]

    
    if (row["absoluteYardlineNumber"] > 10 and row["absoluteYardlineNumber"] <=30 and playDirection == "left") or (row["absoluteYardlineNumber"] >= 90 and row["absoluteYardlineNumber"] < 110 and playDirection == "right"):
        plays.loc[plays['id']==row["id"], ["redZone"]] = True
    else:
        plays.loc[plays['id']==row["id"], ["redZone"]] = False
```
