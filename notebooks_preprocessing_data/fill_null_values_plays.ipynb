Las jugadas con personnelO vacío, van a ser eliminadas porque como se ha visto anteriormente, no coincide en algunas jugadas el personnelO o el personnelD con el número de jugadores en el campo, por tanto tener en cuenta los datos de los jugadores en el campo no sería correcto

```python
plays[plays["personnelO"].isnull()]["id"].values
```

```python
for play in plays[plays["personnelO"].isnull()]["id"].values:
    id_parts = play.split(":")
    week = id_parts[2]
    plays.drop(plays[plays["id"]==play].index,inplace=True)
    weeks[int(week)-1].drop(weeks[int(week)-1][weeks[int(week)-1]["id"]== play].index,inplace=True)
```

Las 17 jugadas que no tienen formación tampoco tiene el número de defensores en la caja. Al ser un número muy limitado, se puede hacer a mano mirando los jugadores con el script.

```python
plays["defendersInTheBox"].unique()
```

```python
file = open("resources/missing_data/formation_defendersbox.txt",'r')
for line in file:
    data = line.split(",")
    playId = data[0]
    formation = data[1]
    defendersInTheBox = data[2][:len(str(data[2]))-1]
    plays.loc[plays['id']==playId, ["offenseFormation"]] = formation
    plays.loc[plays['id']==playId, ["defendersInTheBox"]] = defendersInTheBox
```

Para preSnapVisitorScore, preSnapHomeScore y gameClock, vamos a tener que utilizar un dataset con toda la información (incluido las jugadas eliminadas). Por eso, vamos a utilizar un dataset con todas las jugadas, incluidas las eliminadas, para luego pasar esa información.

```python
all_plays = pd.read_csv("processed_data/new_id/plays_new_id.csv",index_col=0)
plays.head()
```

```python
all_plays[all_plays["preSnapVisitorScore"].isnull()]
```

```python
all_plays["preSnapVisitorScore"].fillna(method="ffill",inplace=True)
all_plays["preSnapHomeScore"].fillna(method="ffill",inplace=True)
all_plays["gameClock"].fillna(method="ffill",inplace=True)
```

Ahora tenemos que pasarlo a nuestro dataset

```python
for index,row in plays[plays["preSnapVisitorScore"].isnull()].iterrows():
    plays.loc[plays['id']==row["id"], ["preSnapVisitorScore"]] = all_plays[all_plays["id"]==row["id"]]["preSnapVisitorScore"].values[0]
    plays.loc[plays['id']==row["id"], ["preSnapHomeScore"]] = all_plays[all_plays["id"]==row["id"]]["preSnapHomeScore"].values[0]
    plays.loc[plays['id']==row["id"], ["gameClock"]] = all_plays[all_plays["id"]==row["id"]]["gameClock"].values[0]
```
