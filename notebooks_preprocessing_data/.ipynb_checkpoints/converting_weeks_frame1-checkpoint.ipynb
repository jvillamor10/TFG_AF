```python
games = pd.read_csv("raw_data/games.csv")
games.head()
```

```python
plays = pd.read_csv("processed_data/clean/plays.csv",index_col=0)
plays.head()
```

```python
offense_players = pd.DataFrame(columns=all_players.columns)
for play in all_players["id"].unique():
    possessionTeam = plays[plays["id"]==play]["possessionTeam"].values[0]
    idParts = play.split(":")
    gameId = idParts[0]
                     
    team = ""
    if possessionTeam == games[games["gameId"]==int(gameId)]["homeTeamAbbr"].values[0]:
        team = "home"
    elif possessionTeam == games[games["gameId"]==int(gameId)]["visitorTeamAbbr"].values[0]:
        team = "away"
    
    players = all_players[(all_players["id"]==play)&(all_players["team"]==team)]
    offense_players = offense_players.append(players)
```

```python
offense_players
```

```python
pickle_file = open('resources/offense_players.pickle','wb')
pickle.dump(offense_players,pickle_file)
pickle_file.close()
```

```python
defense_players = pd.DataFrame(columns=all_players.columns)
for play in all_players["id"].unique():
    possessionTeam = plays[plays["id"]==play]["possessionTeam"].values[0]
    idParts = play.split(":")
    gameId = idParts[0]
                     
    team = ""
    if possessionTeam == games[games["gameId"]==int(gameId)]["homeTeamAbbr"].values[0]:
        team = "away"
    elif possessionTeam == games[games["gameId"]==int(gameId)]["visitorTeamAbbr"].values[0]:
        team = "home"
    
    players = all_players[(all_players["id"]==play)&(all_players["team"]==team)]
    defense_players = defense_players.append(players)
```

```python
defense_players
```

```python
pickle_file = open('resources/defense_players.pickle','wb')
pickle.dump(defense_players,pickle_file)
pickle_file.close()
```

```python
defense_players.to_csv("processed_data/weeks_frame1/frame1_process/defense_players_plays.csv")
offense_players.to_csv("processed_data/weeks_frame1/frame1_process/offense_players_plays.csv")
```
