Vamos a comprobar que esté en plays

```python
plays_football_is_far = plays[plays["id"].isin(football_is_far)]
plays_football_is_far.head()
```

```python
len(plays_football_is_far)
```

```python
for play in plays_football_is_far["id"].values:
    los = plays[plays["id"]==play]["absoluteYardlineNumber"].values[0]
    idParts = play.split(":")
    week = idParts[2]
    weeks[int(week)-1].loc[(weeks[int(week)-1]["id"]==play)&(weeks[int(week)-1]["displayName"] == "Football")&(weeks[int(week)-1]["frameId"]==1),["x"]] = los
```

```python
plays_football_is_outside = plays[plays["id"].isin(football_is_outside)]
plays_football_is_outside.head()
```

```python
len(plays_football_is_outside)
```

```python
for play in plays_football_is_outside["id"].values:
    los = plays[plays["id"]==play]["absoluteYardlineNumber"].values[0]
    idParts = play.split(":")
    week = idParts[2]
    weeks[int(week)-1].loc[(weeks[int(week)-1]["id"]==play)&(weeks[int(week)-1]["displayName"] == "Football")&(weeks[int(week)-1]["frameId"]==1),["x"]] = los
    
    qb_y = weeks[int(week)-1][(weeks[int(week)-1]["id"]==play)&(weeks[int(week)-1]["position"] == 'QB')&(weeks[int(week)-1]["frameId"]==1)]["y"].values[0]
    
    if qb_y >= 23.6 and qb_y <= 29.7: #23.6 y 29.7
        weeks[int(week)-1].loc[(weeks[int(week)-1]["id"]==play)&(weeks[int(week)-1]["displayName"] == "Football")&(weeks[int(week)-1]["frameId"]==1),["y"]] = qb_y
    elif qb_y < 23.6:
        weeks[int(week)-1].loc[(weeks[int(week)-1]["id"]==play)&(weeks[int(week)-1]["displayName"] == "Football")&(weeks[int(week)-1]["frameId"]==1),["y"]] = 23.6
    elif qb_y > 29.7:
        weeks[int(week)-1].loc[(weeks[int(week)-1]["id"]==play)&(weeks[int(week)-1]["displayName"] == "Football")&(weeks[int(week)-1]["frameId"]==1),["y"]] = 29.7
```

```python
week_number = 1
for week in weeks:
    week.to_csv("../processed_data/clean/week{}.csv".format(week_number))
    week_number+=1
```
