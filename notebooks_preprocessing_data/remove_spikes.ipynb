```python
weeks = [week1,week2,week3,week4,week5,week6,week7,week8,week9,week10,week11,week12,week13,week14,week15,week16,week17]
```

```python
pickle_file_spikes = open('resources/spikes_plays.pickle','rb')
spike_plays = pickle.load(pickle_file_spikes)
pickle_file_spikes.close()
```

```python
spike_plays
```

```python
for play in spike_plays:
    id_parts = play.split(":")
    week = id_parts[2]
    plays.drop(plays[plays["id"]==play].index,inplace=True)
    weeks[int(week)-1].drop(weeks[int(week)-1][weeks[int(week)-1]["id"]== play].index,inplace=True)
```

```python
plays
```

```python
for week in weeks:
    week.reset_index(inplace=True)
```

```python
plays.reset_index(inplace=True)
```

```python
week_number = 1
for week in weeks:
    week.to_csv("processed_data/remove_st_nbt_spikes/week{}.csv".format(week_number))
    week_number+=1
```
