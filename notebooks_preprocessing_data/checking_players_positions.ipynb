Calculamos la posición del balón respecto a la línea de Scrimmage

```python
plays_line_far = []
for index,row in plays.iterrows():
    id_parts = row["id"].split(":")
    week = id_parts[2]
    absoluteYardLine = round(weeks_f1[int(week)-1][(weeks_f1[int(week)-1]["id"]==row["id"])&(weeks_f1[int(week)-1]["displayName"]=="Football")]["x"].values[0])
    if abs(absoluteYardLine - int(row["absoluteYardlineNumber"])) > 2:
        plays_line_far.append(row["id"])
```

Algunas jugadas no tienen el valor o aparece fuera del campo cuando es el primer frame. Habrá que ver la posición del balón.

Ahora miramos que ningún jugador sobre pase la línea de scrimmage

```python
players_over_line = []
for week in weeks_f1:
    for index,row in week.iterrows():
        playDirection = row["playDirection"]
        if row["position"] in ['QB','WR','RB','TE','FB','HB']:
            if playDirection == "right":
                if row["x"] > plays[plays["id"]==row["id"]]["absoluteYardlineNumber"].values[0]:
                    players_over_line.append(row["id"])
            if playDirection == "left":
                if row["x"] < plays[plays["id"]==row["id"]]["absoluteYardlineNumber"].values[0]:
                    players_over_line.append(row["id"])
```

```python
pickle_file = open('resources/players_over_line.pickle','wb')
pickle.dump(players_over_line,pickle_file)
pickle_file.close()
```

Vamos a ver ciertas jugadas donde el balón esté muy distanciado del jugador que esté en "perpendicular" (damos un margen de 3 yardas tanto a la derecha como a la izquierda)

```python
football_is_far = []
football_is_outside = []
for week in weeks_f1:
    playsids = week["id"].unique()
    for play in playsids:
        yFootball = int(week[(week["displayName"]=="Football")&(week["id"]==play)]["y"].values[0])
        offensePlayers = week[(week["id"]==play)&(week["position"].isin(['QB','WR','RB','TE','FB','HB']))]
        align = False
        for index,row in offensePlayers.iterrows():
            yPlayer = int(row["y"])
            if abs(yPlayer - yFootball) < 3:
                align = True
                
        if not(align):
            if yFootball > 53.3 or yFootball < 0:
                football_is_outside.append(play)
            else:
                football_is_far.append(play)
```
