```python
pickle_file_no_one_qb = open('resources/no_one_qb.pickle','rb')
pickle_file_repeated_players = open('resources/two_qbs_same.pickle','rb')
pickle_file_not_same_personnelO = open('resources/not_same_personnelO.pickle','rb')
pickle_file_not_same_personnelD = open('resources/not_same_personnelD.pickle','rb')
pickle_file_defense_strange_density = open('resources/defense_strange_density.pickle','rb')
no_one_qb = pickle.load(pickle_file_no_one_qb)
repeated_players = pickle.load(pickle_file_repeated_players)
not_same_personnelO = pickle.load(pickle_file_not_same_personnelO)
not_same_personnelD = pickle.load(pickle_file_not_same_personnelD)
defense_strange_density = pickle.load(pickle_file_defense_strange_density)
pickle_file_no_one_qb.close()
pickle_file_repeated_players.close()
pickle_file_not_same_personnelO.close()
pickle_file_not_same_personnelD.close()
pickle_file_defense_strange_density.close()
```

```python
weeks = [week1,week2,week3,week4,week5,week6,week7,week8,week9,week10,week11,week12,week13,week14,week15,week16,week17]
```

```python
remove_list = []
remove_list += no_one_qb
remove_list += repeated_players
remove_list += not_same_personnelO
remove_list += not_same_personnelD
remove_list += defense_strange_density
```

```python
len(remove_list)
```

```python
remove_list = list(set(remove_list))
```

```python
len(remove_list)
```

```python
for play in remove_list:
    id_parts = play.split(":")
    week = id_parts[2]
    plays.drop(plays[plays["id"]==play].index,inplace=True)
    weeks[int(week)-1].drop(weeks[int(week)-1][weeks[int(week)-1]["id"]== play].index,inplace=True)
```

```python
week_number = 1
for week in weeks:
    week.to_csv("processed_data/remove_st_nbt_spikes_strangeplays/week{}.csv".format(week_number))
    week_number+=1
```
