De 1303 jugadores que tiene el dataset, 240 tiene un guión en medio. Hay que saber cual es la medida y convertirla de manera adecuada.

```python
players[players["height"]=="72"]
```

Pongamos el ejemplo de Nelson Agholor. Mide 1.83 (6 pies). Según el dataset, mide 72 pulgadas, esto es correcto ya que 1 pie = 12 pulgadas. Por tanto, si no hay guión se está indicando la altura en pies.

```python
players[players["height"]=="6-5"]
```

Eli Manning, por ejemplo mide 1.96, 6 pies y 5 pulgadas. 1 pie son 12 pulgadas y una pulgada son 0,0254

```python
players_copy = players.copy()
```

```python
def changeHeight(oldHeight):
    if "-" in oldHeight:
        height = oldHeight.split("-")
        feet = height[0]
        inches = height[1]
        
        value = (int(feet) * 12 + int(inches)) * 0.0254
    else:
        value = int(oldHeight) * 0.0254
        
    value = round(value,2)
    
    return value
```

```python
changeHeight("6-5")
```

```python
players["height"] = players["height"].apply(changeHeight)
```

```python
players
```

```python
def changeWeight(oldWeight):
    return round(oldWeight/2.205)
```

```python
players["weight"] = players["weight"].apply(changeWeight)
```
