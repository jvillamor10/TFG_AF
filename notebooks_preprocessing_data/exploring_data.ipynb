This is because there are some plays which have the same playId

```python
offense_positions = ['QB','WR','RB','TE','FB','HB']
defense_positions = ['SS','FS','MLB','CB','LB','OLB','ILB','DL','DB','NT','S','DT']
```

```python
for week in weeks:
    values = week["id"].unique()
    for play in values:
        positions = week[week["id"]==play]["position"].unique()
        offense = False
        defense = False
        for player in positions:
            if player in offense_positions:
                offense = True
            if player in defense_positions:
                defense = True
        
        if offense != True or defense != True:
            if offense != True:
                print("No offense in play: ",play)
            if defense != True:
                print("No defense in play: ",play)
```
