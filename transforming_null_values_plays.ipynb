Let's start with absoluteYardlineNumber, which is the most important column of those.

```python
plays[plays["absoluteYardlineNumber"].isnull()]
```

To solve this is easy, we have just to obtain the football value of the play in weeks dataframes and round it.

```python
week1 = pd.read_csv("processed_data/weeks_frame1/frame1_raw/week1_frame1.csv")
week2 = pd.read_csv("processed_data/weeks_frame1/frame1_raw/week2_frame1.csv")
week3 = pd.read_csv("processed_data/weeks_frame1/frame1_raw/week3_frame1.csv")
week4 = pd.read_csv("processed_data/weeks_frame1/frame1_raw/week4_frame1.csv")
week5 = pd.read_csv("processed_data/weeks_frame1/frame1_raw/week5_frame1.csv")
week6 = pd.read_csv("processed_data/weeks_frame1/frame1_raw/week6_frame1.csv")
week7 = pd.read_csv("processed_data/weeks_frame1/frame1_raw/week7_frame1.csv")
week8 = pd.read_csv("processed_data/weeks_frame1/frame1_raw/week8_frame1.csv")
week9 = pd.read_csv("processed_data/weeks_frame1/frame1_raw/week9_frame1.csv")
week10 = pd.read_csv("processed_data/weeks_frame1/frame1_raw/week10_frame1.csv")
week11 = pd.read_csv("processed_data/weeks_frame1/frame1_raw/week11_frame1.csv")
week12 = pd.read_csv("processed_data/weeks_frame1/frame1_raw/week12_frame1.csv")
week13 = pd.read_csv("processed_data/weeks_frame1/frame1_raw/week13_frame1.csv")
week14 = pd.read_csv("processed_data/weeks_frame1/frame1_raw/week14_frame1.csv")
week15 = pd.read_csv("processed_data/weeks_frame1/frame1_raw/week15_frame1.csv")
week16 = pd.read_csv("processed_data/weeks_frame1/frame1_raw/week16_frame1.csv")
week17 = pd.read_csv("processed_data/weeks_frame1/frame1_raw/week17_frame1.csv")
week1.head()
```

```python
weeks = [week1,week2,week3,week4,week5,week6,week7,week8,week9,week10,week11,week12,week13,week14,week15,week16,week17]
```

```python
for index,row in plays[plays["absoluteYardlineNumber"].isnull()].iterrows():
    week = row["id"].split(":")
    plays.loc[plays['id']==row["id"], ["absoluteYardlineNumber"]] = round(weeks[int(week[2])-1][(weeks[int(week[2])-1]["id"]==row["id"]) & (weeks[int(week[2])-1]["displayName"]=="Football")]["x"].values[0])
```

The problem is solved.
