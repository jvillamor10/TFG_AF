```python
play_id_dict = {}
```

```python
plays_by_games = []
first = True
for row in example_1:
    if first:
        plays_by_games.append(row[0][1])
        actual_game = row[0][0]
        first = False
    elif row[0][0] == actual_game:
        plays_by_games.append(row[0][1])
    else:
        play_id_dict[actual_game] = plays_by_games
        actual_game = row[0][0]
        plays_by_games = []
        plays_by_games.append(row[0][1])
        
play_id_dict[actual_game] = plays_by_games
        
```
