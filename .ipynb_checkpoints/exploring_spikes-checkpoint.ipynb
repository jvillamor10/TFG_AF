```python
import pandas as pd
import pickle
```

```python
pd.set_option('display.max_columns',None)
```

```python
plays = pd.read_csv("processed_data/remove_st_nbt/plays.csv",index_col=0)
plays.head()
```

```python
list_spike = []
for index,row in plays.iterrows():
    if "spike" in row["playDescription"]:
        list_spike.append(row["id"])
print(len(list_spike))
```

```python
pickle_file = open('resources/spikes_plays.pickle','wb')
pickle.dump(list_spike,pickle_file)
pickle_file.close()
```
