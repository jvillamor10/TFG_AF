```python
plays = pd.read_csv("../processed_data/clean/plays.csv",index_col=1)
plays.head()
```

```python
games = pd.read_csv("../raw_data/games.csv",index_col=0)
games.head()
```

```python
def obtainDefenseTeam(playId):
    playIdParts = playId.split(":")
    gameId = playIdParts[0]
    visitorTeam = games.loc[int(gameId)]["visitorTeamAbbr"]
    homeTeam = games.loc[int(gameId)]["homeTeamAbbr"]
    possessionTeam = plays.loc[playId]["possessionTeam"]
    
    if possessionTeam == visitorTeam: #si equipo visitante tiene el balón, el equipo local es el que defiende
        value = "home"
    elif possessionTeam == homeTeam:
        value = "away"
    
    return value
```

```python
def obtainOffenseTeam(playId):
    playIdParts = playId.split(":")
    gameId = playIdParts[0]
    visitorTeam = games.loc[int(gameId)]["visitorTeamAbbr"]
    homeTeam = games.loc[int(gameId)]["homeTeamAbbr"]
    possessionTeam = plays.loc[playId]["possessionTeam"]
    
    if possessionTeam == visitorTeam: #si equipo visitante tiene el balón, el equipo local es el que defiende
        value = "away"
    elif possessionTeam == homeTeam:
        value = "home"
    
    return value
```

```python
for week in weeks:
    week.drop(["Unnamed: 0","Unnamed: 0.1","Unnamed: 0.1.1","level_0","Unnamed: 0.1.1.1","index"],axis=1,inplace=True)
```

```python
pass_rushers = {}
for week in weeks:
    for play in week["id"].unique():
        players = week[week["id"]==play]
        los = plays.loc[play]["absoluteYardlineNumber"]
        playDirection = players["playDirection"].values[0]
        players_pass_rush = []
        max_frame = -1
        if ('pass_forward' in list(week[week["id"]==play]["event"].unique())):
                max_frame = week[(week["id"]==play)&(week["event"]=='pass_forward')]["frameId"].values[0]

        for frame in range(1,max_frame+1):
            players_frame = players[(players["frameId"]==frame)&(players["team"]==obtainDefenseTeam(play))]
            for index,row in players_frame.iterrows():
                if row["displayName"] == "Football":
                    continue
                if playDirection == "right" and row["x"] < los - 2 and row["nflId"] not in players_pass_rush:
                    players_pass_rush.append(row["nflId"])
                elif playDirection == "left" and row["x"] > los + 2 and row["nflId"] not in players_pass_rush:
                    players_pass_rush.append(row["nflId"])
        pass_rushers[play] = players_pass_rush
```

```python
#save dataset
pickle_file = open('../resources/pass_rushers.pickle','wb')
pickle.dump(pass_rushers,pickle_file)
pickle_file.close()
```

```python
#load dataset
pickle_file = open('../resources/pass_rushers.pickle','rb')
pass_rushers = pickle.load(pickle_file)
pickle_file.close()
```
